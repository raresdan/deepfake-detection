```python
import torch
from diffusers import StableDiffusionXLPipeline

model_id = "stabilityai/stable-diffusion-xl-base-1.0"
pipe = StableDiffusionXLPipeline.from_pretrained(model_id, torch_dtype=torch.float16).to("cuda")

# Define demographics
age_groups = ["young", "middle-aged", "older"]
ethnicities = ["black", "east-asian", "hispanic", "south-asian", "white"]
genders = ["woman", "man"]

# Generate images
num_images_per_prompt = 250
i = 0
while i < num_images_per_prompt:
  for age in age_groups:
      for ethnicity in ethnicities:
          for gender in genders:
              prompt = f"A photo of a {age} {ethnicity} {gender}, portrait, ultra-realistic, DSLR, studio lighting, high quality, 8k, symmetrical face, detailed skin"
              image = pipe(prompt).images[0]
              image.save(f"/content/drive/MyDrive/deepfake/stable_diffusion_xl/image_{i}.png")
              i += 1
```
